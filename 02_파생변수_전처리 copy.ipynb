## 성장성 파생변수 만들기
    총자산증가율, 자기자본증가율, 정상영업손익증가율, 순이익증가율, 매출액증가율

```python
# k-gaap 각 기업의 2010년 재무데이터 추출
last_cap = kgconsep.groupby('회사명', as_index=False)['자산', '자본', '영업손익', '당기순이익', '매출액'].nth(-1).index
last_cap2 = kgconsep.iloc[last_cap, :]


# 2010년 데이터와 k-ifrs 합치기
k = pd.concat([last_cap2, consep])
k.sort_values(by = ['회사명', '회계년도'], ignore_index=True, inplace=True)


# 파생변수 만들기위한 전기 데이터 만들기
k['전기영업손익'] = k['영업손익'].shift().where(k.회사명.eq(k.회사명.shift()))
k['전기순이익'] = k['당기순이익'].shift().where(k.회사명.eq(k.회사명.shift()))
k['전기매출액'] = k['매출액'].shift().where(k.회사명.eq(k.회사명.shift()))
k['전기자산'] = k['자산'].shift().where(k.회사명.eq(k.회사명.shift()))
k['전기자본'] = k['자본'].shift().where(k.회사명.eq(k.회사명.shift()))

k2 = k.copy()

# 컬럼명 재정렬
k2 = k2[['회사명', '거래소코드', '회계년도', '상장일', '상장폐지일', '자산', '전기자산', '자본', '전기자본', 
            '영업손익', '전기영업손익', '당기순이익', '전기순이익', '매출액', '전기매출액']]


# 성장성 재무비율 구하기
k2['총자산증가율'] = ((k2['자산'] - k2['전기자산'])/abs(k2['전기자산']))*100
k2['자기자본증가율'] = ((k2['자본'] - k2['전기자본'])/abs(k2['전기자본']))*100

for i, row in k2.iterrows () :
    if k2.at[i,'전기영업손익'] > 0 or k2.at[i,'전기영업손익'] < 0 and k2.at[i,'영업손익'] < 0:
        k2.at[i,'정상영업손익증가율'] = ((k2.at[i,'영업손익'] - k2.at[i,'전기영업손익'])/abs(k2.at[i,'전기영업손익']))*100
    elif k2.at[i,'전기영업손익'] < 0 and k2.at[i,'영업손익'] > 0:
        k2.at[i,'정상영업손익증가율'] = ((k2.at[i,'영업손익'] - k2.at[i,'전기영업손익'])/abs(k2.at[i,'영업손익']))*100

for i, row in k2.iterrows () :
    if k2.at[i,'전기순이익'] > 0 or k2.at[i,'전기순이익'] < 0 and k2.at[i,'당기순이익'] < 0:
        k2.at[i,'순이익증가율'] = ((k2.at[i,'당기순이익'] - k2.at[i,'전기순이익'])/abs(k2.at[i,'전기순이익']))*100
    elif k2.at[i,'전기순이익'] < 0 and k2.at[i,'당기순이익'] > 0:
        k2.at[i,'순이익증가율'] = ((k2.at[i,'당기순이익'] - k2.at[i,'전기순이익'])/abs(k2.at[i,'당기순이익']))*100

for i, row in k2.iterrows () :
    if k2.at[i,'전기매출액'] > 0 or k2.at[i,'전기매출액'] < 0 and k2.at[i,'매출액'] < 0:
        k2.at[i,'매출액증가율'] = ((k2.at[i,'매출액'] - k2.at[i,'전기매출액'])/abs(k2.at[i,'전기매출액']))*100
    elif k2.at[i,'전기매출액'] < 0 and k2.at[i,'매출액'] > 0:
        k2.at[i,'매출액증가율'] = ((k2.at[i,'매출액'] - k2.at[i,'전기매출액'])/abs(k2.at[i,'매출액']))*100


# 2010년 데이터는 삭제
k3 = k2.copy()

index_to_drop = k3.groupby('회사명', as_index=False)['회사명', '거래소코드', '회계년도', '상장일', '상장폐지일', '자산', '전기자산', '자본', '전기자본',
       '영업손익', '전기영업손익', '당기순이익', '전기순이익', '매출액', '전기매출액', '총자산증가율', '자기자본증가율',
       '정상영업손익증가율', '순이익증가율', '매출액증가율'].nth(0).index
k3.drop(k3.index[index_to_drop], inplace=True)


k4 = k3.copy()

# 회계년도, 상장일 포멧 변경
k4['회계년도'] = pd.to_datetime(k4['회계년도'], format='%Y/%m')
k4['회계년도'] = k4['회계년도'].dt.strftime('%Y%m')

k4['상장일'] = pd.to_datetime(k4['상장일'], format='%Y/%m')
k4['상장일'] = k4['상장일'].dt.strftime('%Y%m')


# 데이터 프레임 저장
k4.to_csv('데이터/1차 전처리/파생변수_성장성.csv', encoding='CP949', index=False)


# 성장성 파생변수 전처리part1과 합치고 내보내기
df = pd.read_csv('데이터/1차 전처리/전처리part1.csv', encoding='CP949')


# 파생변수 컬럼 외 삭제
k5 = k4.copy()

k5.drop(columns=['자산', '전기자산', '자본', '전기자본',
       '영업손익', '전기영업손익', '당기순이익', '전기순이익', '매출액', '전기매출액'], inplace=True)


# merge
merge1 = pd.merge(df, k5, on = ['회사명', '거래소코드', '회계년도', '상장일', '상장폐지일'], how='left')

merge1.to_csv('데이터/1차 전처리/파생변수merge_성장성.csv', encoding='CP949', index=False)
```

    종업원 1인당 부가가치율, 노동장비율, 기계장비율, 자본집약도, 1주당 매출액, 순이익, 현금흐름,순자산, 노동소득분배율, 자기자본구성비율, 재고자산대순운전자본비율

```python
concat = pd.read_csv('데이터/1차 전처리/concat.csv', encoding='CP949')
concat2 = concat.copy()

# 종업원 1인당 부가가치율, 노동장비율, 기계장비율, 자본집약도
concat2['합계_기말인원'] = concat2['합계_기말인원'].fillna(concat2.groupby(['거래소코드', '상장일'])['합계_기말인원'].transform('mean'))
concat2['종업원1인당부가가치율_계산'] = concat2['부가가치(IFRS)(백만원)'] / concat2['합계_기말인원']
concat2['노동장비율_계산'] = concat2['유형자산(요약)(IFRS)(백만원)'] / concat2['합계_기말인원']
concat2['기계장비율_계산'] = ((concat2['기계장치(IFRS)(천원)'] - concat2['(감가상각누계액)(IFRS)(천원)'])/1000) / concat2['합계_기말인원']
concat2['자본집약도_계산'] = concat2['자산(*)(요약)(IFRS)(백만원)'] / concat2['합계_기말인원']


# 1주당 매출액, 순이익, 현금흐름,순자산, 노동소득분배율
concat2['1주당매출액'] = concat2['매출액']/concat2['발행한주식총수']*1000
concat2['1주당순이익'] = concat2['당기순이익']/concat2['발행한주식총수']*1000
concat2['1주당Cashflow'] = concat2['영업활동으로인한현금흐름간접법']/concat2['발행한주식총수']*1000
concat2['1주당순자산'] = concat2['자본(*)(요약)(IFRS)(백만원)']/concat2['발행한주식총수']*1000000
concat2['노동소득분배율'] = concat2['인건비']  / (concat2['법인세비용차감전계속사업손익'] + concat2['인건비'] + concat2['금융비용'] + concat2['임차료'] + concat2['세금과공과'] + concat2['감가상각비']) *100


# 자기자본구성비율, 재고자산 대 순운전자본비율
concat2['자기자본구성비율'] = (concat2['자본']/concat2['자산'])*100
concat2['재고자산대순운전자본비율'] = (concat2['재고자산']/(concat2['유동자산']-concat2['유동부채']))*100


# 전처리 1차 파일 불러오기
df_part1 = pd.read_csv('데이터/1차 전처리/전처리part1.csv', encoding= 'CP949')


# 파생변수 컬럼 외 삭제
concat3 = concat2.copy()

concat3.drop(columns=['영업손익', '매출액', '당기순이익', '자본',
       '이자비용', '이자', '배당금', '영업활동으로인한현금흐름간접법', '발행한주식총수', '자산', '현금및현금성자산',
       '유동부채', '재고자산', '유동자산', '부채', '합계_기말인원', '법인세비용차감전계속사업손익', '인건비',
       '금융비용', '임차료', '세금과공과', '감가상각비', '영업활동으로 인한 현금흐름(간접법)(*)(IFRS)(천원)',
       '자본(*)(요약)(IFRS)(백만원)', '부가가치(IFRS)(백만원)', '유형자산(요약)(IFRS)(백만원)',
       '기계장치(IFRS)(천원)', '(감가상각누계액)(IFRS)(천원)', '자산(*)(요약)(IFRS)(백만원)'], inplace=True)


# merge
merge2 = pd.merge(df_part1, concat3, on = ['회사명', '거래소코드', '회계년도', '상장일', '상장폐지일'], how = 'left')


# 데이터 저장
merge2.to_csv('데이터/1차 전처리/파생변수merge_1차.csv', encoding='CP949', index=False)
```
