### 변수컬럼 제작 과정

```python
# 변수생성을 위한 재료데이터프레임 불러오기
import pandas as pd
ts = pd.read_csv('./전처리/concat.csv',encoding='cp949')
ts
```

### 1주당 매출액, 순이익, 현금흐름,순자산, 노동소득분배율 컬럼 제작

```python
ts['1주당매출액'] = ts['매출액']/ts['발행한주식총수']*1000
ts['1주당순이익'] = ts['당기순이익']/ts['발행한주식총수']*1000
ts['1주당Cashflow'] = ts['영업활동으로인한현금흐름간접법']/ts['발행한주식총수']*1000
ts['1주당순자산'] = ts['자본(*)(요약)(IFRS)(백만원)']/ts['발행한주식총수']*1000000
ts['노동소득분배율'] = ts['인건비']  / (ts['법인세비용차감전계속사업손익'] + ts['인건비'] + ts['금융비용'] + ts['임차료'] + ts['세금과공과'] + ts['감가상각비']) *100
```

```python
# 분자와 분모, 컬럼의 0과 결측치 파악
#분모
print('분모')
print(len(ts[ts['발행한주식총수']== 0 ]),ts['발행한주식총수'].isna().sum())
#주당 매출액
print('주당 매출액')
print(len(ts[ts['매출액']== 0 ]),ts['매출액'].isna().sum())
print(len(ts[ts['1주당매출액']== 0 ]),ts['1주당매출액'].isna().sum())
#주당 순이익
print('순이익')
print(len(ts[ts['당기순이익']== 0 ]),ts['당기순이익'].isna().sum())
print(len(ts[ts['1주당순이익']== 0 ]),ts['1주당순이익'].isna().sum())
#주당 현금흐름
print('현금흐름')
print(len(ts[ts['영업활동으로인한현금흐름간접법']== 0 ]),ts['영업활동으로인한현금흐름간접법'].isna().sum())
print(len(ts[ts['1주당Cashflow']== 0 ]),ts['1주당Cashflow'].isna().sum())
#주당 순자산
print('순자산')
print(len(ts[ts['자본(*)(요약)(IFRS)(백만원)']== 0 ]),ts['자본(*)(요약)(IFRS)(백만원)'].isna().sum())
print(len(ts[ts['1주당순자산']== 0 ]),ts['1주당순자산'].isna().sum())
#노동소득분배율
print('노동소득분배율')
print(len(ts[ts['노동소득분배율']== 0 ]),ts['노동소득분배율'].isna().sum())
```
