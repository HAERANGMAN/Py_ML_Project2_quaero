생산성 지표 계산식 작성

```python
## 계산식
# 종업원 1인당 부가가치율 = 부가가치(개별 재무제표 이용) / 직원 수 (합계_기말인원)
# 노동장비율 = 유형자산(개별 재무제표) / 직원 수
# 기계장비율 = (기계장치 - 감가상각누계액) / 직원 수
# 자본집약도 = 총자산(개별 재무제표) / 직원 수


df['종업원1인당부가가치율_계산'] = df['부가가치(IFRS)(백만원)'] / df['합계_기말인원']
df['노동장비율_계산'] = df['유형자산(요약)(IFRS)(백만원)'] / df['합계_기말인원']
df['기계장비율_계산'] = ((df['기계장치(IFRS)(천원)'] - df['(감가상각누계액)(IFRS)(천원)'])/1000) / df['합계_기말인원']
df['자본집약도_계산'] = df['자산(*)(요약)(IFRS)(백만원)'] / df['합계_기말인원']
```

```python
# 문제되는 회사 - 생산성지표 작성 후 결과 확인

df[['회사명', '합계_기말인원', '종업원1인당부가가치율_계산', '노동장비율_계산', '기계장비율_계산', '자본집약도_계산']]\
   [(df['거래소코드'] == 192080)| 
    (df['거래소코드'] == 26960) | 
    (df['거래소코드'] == 68270) | 
    (df['거래소코드'] == 35510) | 
    (df['거래소코드'] == 35720) | 
    (df['거래소코드'] == 3670)]
```

# 전처리 part1 파일에 merge

```python
# 파일 불러오기 : 1차 전처리 파일 - 스팩주 등등 외 중복 거래소코드 삭제 ver.

df_part1 = pd.read_csv('data/전처리part1.csv', encoding= 'CP949')
df_part1.drop(columns = 'Unnamed: 0', inplace = True)
print('shape : ', df_part1.shape)
df_part1.head()
```

```python
# 전처리 파일 개수 확인
print('전처리 part1 거래소코드 개수 :', df_part1['거래소코드'].nunique())
print('전처리 part1 회사명 개수 :', df_part1['회사명'].nunique())
print('전처리 part1 shape :', df_part1.shape)

# 계산파일 개수 확인
print('\n계산파일 거래소코드 개수 :', df['거래소코드'].nunique())
print('계산파일 shape :', df.shape)
```

```python
# merge
df_merge = pd.merge(df_part1, df, on = ['거래소코드', '회계년도', '상장일', '상장폐지일'], how = 'left')

print('merge 파일 shape :', df_merge.shape)
print('merge 파일 거래소코드 개수 :', df_merge['거래소코드'].nunique())
print('merge 파일 회사명 개수 :', df_merge['회사명_x'].nunique())
df_merge.head()
```
